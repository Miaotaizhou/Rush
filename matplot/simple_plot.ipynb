```python
%matplotlib notebook
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
```

```python
x=np.arange(0.,10,0.2)
y1=np.cos(x)
y2=np.sin(x)
y3=np.sqrt(x)
```

```python
plt.rcParams['figure.figsize']=(12,8) # 设置图片尺寸
plt.plot(x,y1,color='blue',linewidth=1.5,linestyle='-',marker='.',label=r'$y=cos{x}$')
plt.plot(x,y2,color='green',linewidth=1.5,linestyle='-',marker='*',label=r'$y=sin{x}$')
plt.plot(x,y3,color='m',linewidth=1.5,linestyle='-',marker='x',label=r'$y=\sqrt{x}$')
```

```python
ax = plt.subplot(111) # 设置一个空图
ax.spines['right'].set_color('none') # 设置右边轴线为透明色
ax.spines['top'].set_color('none')
```

```python
'''移动坐标轴的位置'''
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0.1))
```

```python
x=np.arange(0.,10,0.2)
plt.xlim(x.min()*(1.1),x.max()*(1.1))
plt.ylim(-1.5,4.0)
plt.xticks([2,4,6,8,10],[r'two',r'four',r'six',r'eight',r'ten'])
plt.yticks([-1.0,0.0,1.0,2.0,3.0,4.0],[r'bottom',r'0.0',r'1.0',r'2.0',r'3.0',r'4.0'])
plt.title(r'$the \ function \ figure \ of \ cos(),\ sin() \ and \ sqrt()$',fontsize=19)
plt.xlabel(r'$the \ input \ value \ of \ x$',fontsize=18,labelpad=18.8)
plt.ylabel(r'$y=f(x)$',fontsize=18,labelpad=12.5)
```

```python
x=np.arange(0.,10,0.2)
plt.xlim(x.min()*(1.1),x.max()*(1.1))
plt.ylim(-1.5,4.0)
plt.scatter([8,],[np.sqrt(8),],50,color='m')
plt.annotate(r'$2\sqrt{2}$',xy=(8,np.sqrt(8)),xytext=(8.5,2.2),fontsize=16,color='#090909',\
             arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=0.1',color='#090909'))
# xy参数设置'被注解点'的坐标,xytext参数设置'注解文字'的位置,arrowprops参数设置注解文字与被注解点的连接方式
```

```python
# 曲线图
x=np.arange(-5,5,0.1)
y=x**2
plt.plot(x,y)
plt.grid(True)
plt.show()
```

```python
# 直方图
x=np.random.normal(size=1000)
plt.hist(x,bins=10)  #bins参数设置分桶数目
```

```python
# 散点图
plt.rcParams['figure.figsize']=(8,8)
x=np.random.normal(size=1000)
y=np.random.normal(size=1000)
plt.scatter(x,y)
```

```python
# 箱图
x=np.arange(-5,5,0.1)
plt.boxplot(x)
```

```python
# 饼图
sort=['Foxes','Cats','Dogs','Pigs']
sizes=[13,30,30,37]
color=['m','b','g','r']
explode=(0,0.1,0,0)   #分离第2类,数值表示分离远近
plt.pie(sizes,explode=explode,labels=sort,colors=color,autopct='%1.1f%%',shadow=True)  
#autopct参数加百分比
```
